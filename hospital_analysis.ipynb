```python
df = pd.read_csv(r"C:\Users\parth\Downloads\services_weekly.csv")

df['week'] = pd.to_datetime(df['week'], unit='W', origin='2020-01-01')

df = df.sort_values('week')
df.set_index('week', inplace=True)

df.head()
```

```python
weekly_df = df.groupby('week')['patients_request'].sum().reset_index()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(weekly_df['week'], weekly_df['patients_request'])

plt.title("Weekly Patient Demand")
plt.xlabel("Week")
plt.ylabel("Total Demand")

plt.show()
```

```python
weekly_df['forecast'] = weekly_df['patients_request'].rolling(window=4).mean()
```

```python
plt.figure(figsize=(10,5))

plt.plot(weekly_df['week'], weekly_df['patients_request'], label='Actual')
plt.plot(weekly_df['week'], weekly_df['forecast'], label='Forecast')

plt.legend()
plt.title("Demand Forecasting")

plt.show()
```

```python
next_week = weekly_df['patients_request'].tail(4).mean()

print("Predicted Demand for Next Week:", next_week)
```
